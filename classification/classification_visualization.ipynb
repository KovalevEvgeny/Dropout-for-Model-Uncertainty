```python
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
plt.figure(figsize=(12, 8))

LOG_DIR = 'logs'
train_loss_05 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-5_train_loss.npy'))
train_loss_025 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-25_train_loss.npy'))
train_loss_075 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-75_train_loss.npy'))

plt.plot(train_loss_05, label='LeNet, p=0.5')
plt.plot(train_loss_025, label='LeNet, p=0.25')
plt.plot(train_loss_075, label='LeNet, p=0.75')
plt.legend(loc='best')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

LOG_DIR = 'logs'
train_accuracy_05 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-5_train_accuracy.npy'))
train_accuracy_025 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-25_train_accuracy.npy'))
train_accuracy_075 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-75_train_accuracy.npy'))

plt.plot(train_accuracy_05, label='LeNet, p=0.5')
plt.plot(train_accuracy_025, label='LeNet, p=0.25')
plt.plot(train_accuracy_075, label='LeNet, p=0.75')
plt.legend(loc='best')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

LOG_DIR = 'logs'
test_loss_05 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-5_test_loss.npy'))
test_loss_025 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-25_test_loss.npy'))
test_loss_075 = np.load(os.path.join(LOG_DIR, 'lenet_dropout_0-75_test_loss.npy'))

plt.plot(test_loss_05, label='LeNet, p=0.5')
plt.plot(test_loss_025, label='LeNet, p=0.25')
plt.plot(test_loss_075, label='LeNet, p=0.75')
plt.legend(loc='best')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```
